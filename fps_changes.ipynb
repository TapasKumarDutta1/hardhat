```python
import cv2
import os
from tqdm import tqdm
image_folder = '/content/gdrive/MyDrive/predictions_1'
video_name = 'video_5000.avi'

images = [img for img in os.listdir(image_folder) if img.endswith(".jpg")]
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

video = cv2.VideoWriter(video_name, 0, 1, (height,width))

for i in tqdm(range(len(images))):
    video.write(cv2.imread( '/content/gdrive/MyDrive/predictions_1/hardhat_'+str(i)+'.jpg'))

cv2.destroyAllWindows()
video.release()
```
